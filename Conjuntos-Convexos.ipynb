## Conjuntos Convexos

### Norma Euclídea(Norma 2) en $\mathbb{R}^n$

Sea $x = (x_1, x_2, \dots, x_n) \in \mathbb{R}^n$. La norma de $x$ está dada por:

$$
\|x\| = \sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}
$$

### Distancia Euclídea

Sean $x, y \in \mathbb{R}^n$. Definimos la distancia entre $x$ e $y$ por:

$$
d(x, y) = \|x - y\|
$$

Es decir:

$$
d(x, y) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2 + \cdots + (x_n - y_n)^2}
$$

### Bola Abierta

Sean $x_0 \in \mathbb{R}^n$ y $r \geq 0$. La **bola abierta** de centro $x_0$ y radio $r$ es el conjunto:

$$
B(x_0, r) = \{ \, x \in \mathbb{R}^n : \|x - x_0\| < r \, \}
$$

### Conjunto Acotado

Sea $A$ un subconjunto de $\mathbb{R}^n$. Se dice que $A$ es **acotado** cuando existe $r > 0$ tal que:

$$
A \subset B(x_0, r)
$$

Es decir, todo el conjunto $A$ puede quedar contenido dentro de una bola de radio finito.

### Punto Interior

Sea $A \subset \mathbb{R}^n$ y $x \in \mathbb{R}^n$. Decimos que $x$ es un **punto interior** de $A$ si existe $r > 0$ tal que:

$$
B(x, r) \subset A
$$

El conjunto de todos los puntos interiores de $A$ se denota por:

$$
\mathrm{int}(A)
$$

### Conjunto Abierto y Conjunto Cerrado

Sea $A \subset \mathbb{R}^n$. Se dice que $A$ es **abierto** si:

$$
A = \mathrm{int}(A)
$$

Es decir, todo punto de $A$ es un punto interior de $A$.

Un conjunto C es cerrado si su complemento es un conjunto abierto.

### Segmento Lineal

Dados $x, y \in \mathbb{R}^n$:

- El **segmento cerrado** entre $x$ e $y$ es:

$$
[x,y] = \{ \, tx + (1-t)y \in \mathbb{R}^n : 0 \leq t \leq 1 \, \}
$$

- El **segmento abierto** entre $x$ e $y$ es:

$$
]x,y[ = \{ \, tx + (1-t)y \in \mathbb{R}^n : 0 < t < 1 \, \}
$$

### Conjunto Compacto

Un conjunto $C \subset \mathbb{R}^n$ es compacto si es un conjunto cerrado y acotado.

### Frontera de un Conjunto

Sea $A \subset \mathbb{R}^n$. La **frontera** de $A$ es el conjunto:

$$
\mathrm{front}(A) = \{ \, x \in \mathbb{R}^n : B(x,\varepsilon)\cap A \neq \varnothing, \;
B(x,\varepsilon)\cap A^c \neq \varnothing \;\; \text{ para todo } \varepsilon > 0 \, \}
$$

Es decir, un punto $x$ pertenece a la frontera de $A$ si **cada bola** alrededor de $x$ intersecta tanto a $A$ como a su 
complemento $A^c$.

### Conjuntos Convexos

Un conjunto $\Omega \subset \mathbb{R}^n$ es **convexo** si y solo si:

$$
\forall x, y \in \Omega, \;\; [x,y] \subset \Omega
$$

donde el **segmento cerrado** entre $x$ y $y$ está dado por:

$$
[x,y] = \{\, tx + (1-t)y \in \mathbb{R}^n : 0 \leq t \leq 1 \,\}.
$$

### Teorema : Intersección de conjuntos convexos

Sean $S_i \subset \mathbb{R}^n$, $i = 1,2,\dots,n$, conjuntos convexos arbitrarios. Entonces su intersección finita

$$
S = \bigcap_{i=1}^{n} S_i
$$

es también un conjunto **convexo**.

### Propiedades de la convexidad

Sean $\Omega_1, \Omega_2 \subset \mathbb{R}^n$ conjuntos convexos.

1. **Intersección**  
   $$
   \Omega_1 \cap \Omega_2 \;\; \text{es convexo.}
   $$

2. **Suma de Minkowski**  
   $$
   \Omega_1 + \Omega_2 = \{\, x + y \in \mathbb{R}^n : x \in \Omega_1, \; y \in \Omega_2 \,\}
   $$  
   es un conjunto convexo.

3. **Resta de Minkowski**  
   $$
   \Omega_1 - \Omega_2 = \{\, x - y \in \mathbb{R}^n : x \in \Omega_1, \; y \in \Omega_2 \,\}
   $$  
   es un conjunto convexo.

### Combinación lineal

Dado $x \in \mathbb{R}^n$ y $C \subset \mathbb{R}^n$, decimos que $x$ es una **combinación lineal** de elementos de $C$ si existen:

- $p \in \mathbb{N}$  
- escalares $\{t_i\}_{i=1}^p \in \mathbb{R}$  
- vectores $\{x_i\}_{i=1}^p \subset C$  

tales que

$$
x = \sum_{i=1}^{p} t_i x_i.
$$

### Combinación afín

Dado $x \in \mathbb{R}^n$ y $C \subset \mathbb{R}^n$, decimos que $x$ es una **combinación afín** de elementos de $C$ si existen:

- $p \in \mathbb{N}$  
- escalares $\{t_i\}_{i=1}^p \subset \mathbb{R}$  
- vectores $\{x_i\}_{i=1}^p \subset C$  

tales que

$$
x = \sum_{i=1}^p t_i x_i
\quad \text{y} \quad
\sum_{i=1}^p t_i = 1.
$$

### Combinación convexa

Dado $x \in \mathbb{R}^n$ y $C \subset \mathbb{R}^n$, decimos que $x$ es una **combinación convexa** de elementos de $C$ si existen:

- $p \in \mathbb{N}$  
- escalares $\{t_i\}_{i=1}^p \subset [0,1]$  
- vectores $\{x_i\}_{i=1}^p \subset C$  

tales que

$$
x = \sum_{i=1}^p t_i x_i,
\quad \text{con} \quad
\sum_{i=1}^p t_i = 1.
$$

### Cápsula convexa

Sea $C \subset \mathbb{R}^n$. La **cápsula convexa** de $C$ es el conjunto de todas las combinaciones convexas de elementos de $C$.  

Se denota por:

$$
\operatorname{co}(C) =
\left\{
\sum_{i=1}^p t_i x_i \;\middle|\;
p \in \mathbb{N}, \;
t_i \in [0,1], \;
\sum_{i=1}^p t_i = 1, \;
x_i \in C
\right\}.
$$

### Teorema: Carácterización por la cápsula convexa

Sea $C \subset \mathbb{R}^n$. Entonces $C$ es convexo **si y solo si** se cumple

$$
C = \operatorname{co}(C).
$$

### Hiperplano

Sea $a \in \mathbb{R}^n$ con $a \neq 0$ y $b \in \mathbb{R}$. Un **hiperplano** $H \subset \mathbb{R}^n$ es el conjunto definido como:

$$
H = \{ \, x \in \mathbb{R}^n : a^T x = b \, \}
$$

El vector $a$ es un **vector normal** al hiperplano $H$,  
y $a^T x$ es un **producto escalar**:

$$
a^T x = a_1 x_1 + a_2 x_2 + \cdots + a_n x_n \in \mathbb{R}.
$$

### Semiespacios

Sea $a \in \mathbb{R}^n$ con $a \neq 0$ y $b \in \mathbb{R}$. Un **semiespacio** $S \subset \mathbb{R}^n$ es un conjunto definido como:

$$
S = \{ \, x \in \mathbb{R}^n : a^T x \leq b \, \}
$$

o también como:

$$
S = \{ \, x \in \mathbb{R}^n : a^T x \geq b \, \}.
$$

### Poliedro

Un **poliedro** es la **intersección finita de semiespacios**. Todo poliedro es un conjunto **convexo**.

**Ejemplo en** $\mathbb{R}^2$:

Sea
$$
S_1 = \{ x \in \mathbb{R}^2 : x_1 + 3x_2 \leq 7 \},
$$

$$
S_2 = \{ x \in \mathbb{R}^2 : 2x_1 - 3x_2 \leq 2 \},
$$

$$
S_3 = \{ x \in \mathbb{R}^2 : 2x_1 + x_2 \geq 3 \}.
$$

Entonces el poliedro $P$ está dado por:

$$
P = S_1 \cap S_2 \cap S_3.
$$

### Punto Extremo

Sea $\Omega \subset \mathbb{R}^n$ un conjunto convexo no vacío.  
Se dice que el punto $x \in \Omega$ es un **vértice** o **punto extremo** de $\Omega$ si y solo si:

$$
x = t x_1 + (1 - t)x_2 \quad \text{para algún } t \in [0,1], \; x_1, x_2 \in \Omega 
\;\;\; \Rightarrow \;\;\; x_1 = x_2 = x.
$$

Ejemplos:

- En **$\mathbb{R}^1$** (la recta real), los **puntos extremos** de un intervalo cerrado 
  \([a,b]\) son precisamente \(a\) y \(b\).

- En **$\mathbb{R}^2$**, si tienes un polígono convexo (por ejemplo, un triángulo, cuadrado, hexágono, etc.), 
  los **puntos extremos** son exactamente sus **vértices**.

- En **$\mathbb{R}^3$**, para un poliedro convexo (como un cubo o un tetraedro), 
  los **puntos extremos** son los **vértices del sólido**.
