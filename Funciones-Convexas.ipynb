## Funciones Convexas

### Función Convexa

Sea $f : \Omega \subset \mathbb{R}^n \to \mathbb{R}$, con $\Omega$ un conjunto **convexo** no vacío.  
Decimos que $f$ es **convexa sobre** $\Omega$ si y solo si:

$$
\forall x, y \in \Omega, \; \forall t \in [0,1] \quad 
f(tx + (1 - t)y) \leq t f(x) + (1 - t)f(y).
$$

Además, $f$ es **estrictamente convexa** si:

$$
\forall x, y \in \Omega, \; x \neq y, \; \forall t \in (0,1) \quad 
f(tx + (1 - t)y) < t f(x) + (1 - t)f(y).
$$

### Función Cóncava

Sea $f : \Omega \subset \mathbb{R}^n \to \mathbb{R}$, con $\Omega$ un conjunto **convexo** no vacío.  
Decimos que $f$ es **cóncava sobre** $\Omega$ si y solo si:

$$
\forall x, y \in \Omega, \; \forall t \in [0,1] \quad 
f(tx + (1 - t)y) \geq t f(x) + (1 - t)f(y).
$$

Además, $f$ es **estrictamente cóncava** si:

$$
\forall x, y \in \Omega, \; x \neq y, \; \forall t \in (0,1) \quad 
f(tx + (1 - t)y) > t f(x) + (1 - t)f(y).
$$

### Teorema 1

Si $A$ es un subconjunto **convexo** y no vacío de $\mathbb{R}^n$,  
y $f : A \to \mathbb{R}$ es una función **convexa**,  
entonces $f$ es **continua en el interior de $A$**.

### Epígrafo

Sea $X$ un subconjunto no vacío de $\mathbb{R}^n$ y $f : X \to \mathbb{R}$ una función.  
El **epígrafo** de la función $f$ es el conjunto:

$$
\operatorname{epi}(f) = \{(x, y) \in \mathbb{R}^n \times \mathbb{R} : x \in X,\; y \geq f(x)\}
$$

### Hipógrafo

Sea $X$ un subconjunto no vacío de $\mathbb{R}^n$ y $f : X \to \mathbb{R}$ una función.  
El **hipógrafo** de la función $f$ es el conjunto:

$$
\operatorname{hip}(f) = \{(x, y) \in \mathbb{R}^n \times \mathbb{R} : x \in X,\; y \leq f(x)\}
$$

### Teorema 2

Sean $C$ un subconjunto convexo de $\mathbb{R}^n$ y $f : C \to \mathbb{R}$ una función.  

Se cumple que **$f$ es convexa si y solo si su epígrafo es un conjunto convexo en $\mathbb{R}^n \times \mathbb{R}$.**

### Función Diferenciable

Sea $f : \mathbb{R}^n \to \mathbb{R}$.  
Decimos que $f$ es **diferenciable en $x_0$** si existen las siguientes $n$ derivadas parciales:

$$
\frac{\partial f}{\partial x_i}(x_0), \quad i = 1, 2, \dots, n
$$

y además se cumple que:

$$
\lim_{x \to x_0} 
\frac{\; | f(x) - f(x_0) - \nabla f(x_0)^T \cdot (x - x_0) | \;}
{\| x - x_0 \|} = 0
$$

donde el **gradiente de $f$ en $x_0$** está dado por:

$$
\nabla f(x_0)^T =
\left(
\frac{\partial f}{\partial x_1}(x_0),
\;
\dots,
\;
\frac{\partial f}{\partial x_n}(x_0)
\right).
$$

### Teorema 3

Sea el conjunto $S \subset \mathbb{R}^n$ convexo y no vacío, y la función $f : \mathbb{R}^n \to \mathbb{R}$ diferenciable sobre $S$.  

Entonces, $f$ es **convexa sobre $S$** si y solo si para cualesquiera $x, x_0 \in S$ se cumple:

$$
f(x) \;\;\geq\;\; f(x_0) + \nabla f(x_0)^T (x - x_0).
$$

### Funciones dos Veces Diferenciables

Sean $S$ un conjunto no vacío en $\mathbb{R}^n$ y la función $f : S \to \mathbb{R}$.  

Decimos que $f$ es **dos veces diferenciable** en $x_0 \in S$ si existen:

- el vector gradiente $\nabla f(x_0)$,  
- la matriz Hessiana $H(x_0)$,  
- y una función $g : \mathbb{R}^n \to \mathbb{R}$  

tales que:

$$
f(x) \;=\; f(x_0) + \nabla f(x_0)^T (x - x_0) \;+\; \tfrac{1}{2} (x - x_0)^T H(x_0) (x - x_0) \;+\; \|x - x_0\|^2 g(x)
$$

para cada $x \in S$, donde:

$$
\lim_{x \to x_0} g(x) = 0.
$$

- Hasta el **segundo término** se dice **aproximación lineal**.  
- Hasta el **tercer término** se dice **aproximación cuadrática**.

### Matriz Hessiana

Sean $S$ un conjunto no vacío en $\mathbb{R}^n$ y la función $f : S \to \mathbb{R}$. Sea $x_0 \in \mathbb{R}^n$.

La **matriz Hessiana** de $f$ en $x_0$ se denota por $H(x_0)$ y está dada por:

$$
H(x_0) =
\begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2}(x_0) & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n}(x_0) \\
\vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n \partial x_1}(x_0) & \cdots & \frac{\partial^2 f}{\partial x_n^2}(x_0)
\end{bmatrix}
$$

#### Propiedades de la Hessiana

1. **Semidefinida positiva (SDP):**  
   $$
   x^T H(x_0) x \;\geq\; 0 \quad \forall x \in \mathbb{R}^n
   $$

2. **Semidefinida negativa (SDN):**  
   $$
   x^T H(x_0) x \;\leq\; 0 \quad \forall x \in \mathbb{R}^n
   $$

3. **Indefinida:**  
   Si no es ni semidefinida positiva ni semidefinida negativa.

### Teorema 4: Convexidad y Hessiana

Sea $S$ un conjunto convexo no vacío en $\mathbb{R}^n$, y sea $f : S \to \mathbb{R}$ una función dos veces diferenciable en $S$.

- $f$ es **convexa** si y solamente si su **matriz Hessiana** es **semidefinida positiva** en cada punto de $S$:

$$
H(x) \succeq 0 \quad \forall x \in S
$$

- $f$ es **cóncava** si y solamente si su **matriz Hessiana** es **semidefinida negativa** en cada punto de $S$:

$$
H(x) \preceq 0 \quad \forall x \in S
$$

### Mínimos y Máximos

#### Mínimo Global
Un punto $x_0 \in \Omega$ es un **mínimo global** de $f$ si:

$$
\forall x \in \Omega, \quad f(x_0) \leq f(x)
$$

#### Máximo Global
Un punto $x_0 \in \Omega$ es un **máximo global** de $f$ si:

$$
\forall x \in \Omega, \quad f(x_0) \geq f(x)
$$

#### Mínimo Local
Un punto $x_0 \in \Omega$ es un **mínimo local** de $f$ si:

$$
\exists \, \varepsilon > 0 \;\; \text{tal que} \;\; \forall x \in \Omega, \; 
\|x - x_0\| < \varepsilon \;\; \Rightarrow \;\; f(x_0) \leq f(x)
$$

#### Máximo Local
Un punto $x_0 \in \Omega$ es un **máximo local** de $f$ si:

$$
\exists \, \varepsilon > 0 \;\; \text{tal que} \;\; \forall x \in \Omega, \; 
\|x - x_0\| < \varepsilon \;\; \Rightarrow \;\; f(x_0) \geq f(x)
$$

### Teorema: Condición de óptimo para funciones convexas

Sea $f : S \to \mathbb{R}$ una función convexa definida sobre un conjunto convexo no vacío 
$S \subset \mathbb{R}^n$.  

Si se cumple que:

$$
\nabla f(x_0) = 0,
$$

entonces $x_0$ es el **único mínimo** de $f$.

### Teorema: Extremos en poliedros compactos

Sea $f : S \to \mathbb{R}$ una función convexa definida sobre un conjunto convexo no vacío $S \subset \mathbb{R}^n$.  

El **máximo o mínimo** de $f$ sobre un **poliedro compacto no vacío** 
$X \subset S$ se alcanza en un **punto frontera** de $X$.
