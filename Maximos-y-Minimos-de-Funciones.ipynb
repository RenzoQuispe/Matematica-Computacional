# Máximos y Mínimos de Funciones

### <u>El espacio $\mathbb{R}^n$</u>

1. $\mathbb{R}^1$ es la **recta real**, cuyos elementos son los números reales.

2. $\mathbb{R}^2$ es el **plano**, cuyos elementos son los pares ordenados $(x, y)$.

3. $\mathbb{R}^3$ es el **espacio tridimensional**, cuyos elementos son las ternas $(x, y, z)$.

4. En general, $\mathbb{R}^n$ es el **espacio $n$-dimensional**, cuyos elementos son $n$-uplas ordenadas:

$$
(x_1, x_2, \dots, x_n), \quad x_i \in \mathbb{R}
$$

### <u>Funciones de Varias Variables</u>

#### 1. Función de una variable

Una función de una variable es una aplicación definida en un subconjunto de $\mathbb{R}$.

**Ejemplo:**

$$
f : [1, 3] \to \mathbb{R}, 
\quad f(x) = x^2
$$

#### 2. Función de dos variables

Una función de dos variables es una aplicación definida en un subconjunto de $\mathbb{R}^2$.

**Ejemplo:**

$$
f : A \to \mathbb{R}, 
\quad A = \{ (x, y) \in \mathbb{R}^2 : xy > 0 \}, 
\quad f(x, y) = x^2 + xy + y^2
$$

#### 3. Función de tres variables

Una función de tres variables es una aplicación definida en un subconjunto de $\mathbb{R}^3$.

**Ejemplo:**

$$
f : \mathbb{R}^3 \to \mathbb{R}, 
\quad f(x, y, z) = xy - z
$$

### <u>Representación Gráfica de Funciones de Dos Variables</u>

Sea $f : A \subseteq \mathbb{R}^2 \to \mathbb{R}$ una función de dos variables.  

La **gráfica** de $f$, denotada por $\operatorname{gr}(f)$, es el conjunto:

$$
\operatorname{gr}(f) 
= \{ (x, y, z) \in \mathbb{R}^3 : (x, y) \in A, \; z = f(x, y) \}.
$$

Es decir, la gráfica de una función de dos variables está formada por los puntos en el **espacio tridimensional** $(x, y, z)$, donde el valor de $z$ corresponde a la evaluación de la función $f(x, y)$.

### <u>Convexidad</u>

Sea $f : I \to \mathbb{R}$ una función definida en un intervalo $I$.

#### Teorema: Criterios de Convexidad

Son **equivalentes** las siguientes condiciones:

1. $f$ es **convexa** en $I$.  
2. La derivada $f'(x)$ es **creciente** en $I$.  
3. La segunda derivada cumple:  
   $$
   f''(x) \;\geq\; 0, \quad \forall x \in \operatorname{int}(I).
   $$

Ejemplos:

1. $f(x) = x^2$  
   $f''(x) = 2 \;\geq\; 0 \;\;\; \Rightarrow \;\;$ $f$ es convexa en $\mathbb{R}$.  

2. $f(x) = e^x$  
   $f''(x) = e^x > 0 \;\;\; \Rightarrow \;\;$ $f$ es convexa en $\mathbb{R}$.  

### <u>Concavidad</u>

Sea $f : I \to \mathbb{R}$ una función definida en un intervalo $I$.

#### Teorema: Criterios de Concavidad

Son **equivalentes** las siguientes condiciones:

1. $f$ es **cóncava** en $I$ (equivalentemente, $-f$ es convexa).  
2. La derivada $f'(x)$ es **decreciente** en $I$.  
3. La segunda derivada cumple:  
   $$
   f''(x) \;\leq\; 0, \quad \forall x \in \operatorname{int}(I).
   $$

Ejemplos:

1. $f(x) = -x^2$  
   $f''(x) = -2 \;\leq\; 0 \;\;\; \Rightarrow \;\;$ $f$ es cóncava en $\mathbb{R}$.  

2. $f(x) = \ln(x)$, definida en $(0, \infty)$  
   $f''(x) = -\tfrac{1}{x^2} \;\leq\; 0 \;\;\; \Rightarrow \;\;$ $f$ es cóncava en $(0, \infty)$.  

### <u>Puntos de Inflexión</u>

#### Definición
Sea $f : I \to \mathbb{R}$ derivable en el interior de $I$.  
Decimos que $c \in I$ es un **punto de inflexión** de $f$ si la función presenta un **cambio de concavidad** en $c$ (de convexa a cóncava o viceversa).

#### Teorema
Si $f$ es **dos veces derivable** y $c$ es un punto de inflexión de $f$,  
entonces:

$$
f''(c) = 0
$$

Observación:

El **recíproco no siempre se cumple**:  
- Que $f''(c) = 0$ **no garantiza** que $c$ sea un punto de inflexión.  
- Es necesario además verificar que **exista un cambio de signo en $f''(x)$** alrededor de $c$.  

### <u>Máximos y Mínimos</u>

Sea $f : I \to \mathbb{R}$ una función.

1. **Mínimo absoluto:**  
   Si 
   $$
   f(x_0) \leq f(x), \quad \forall x \in I,
   $$
   entonces $x_0$ es un **mínimo absoluto**.

2. **Máximo absoluto:**  
   Si 
   $$
   f(x_0) \geq f(x), \quad \forall x \in I,
   $$
   entonces $x_0$ es un **máximo absoluto**.

3. **Mínimo local:**  
   Si 
   $$
   f(x_0) \leq f(x), \quad \forall x \text{ cercano a } x_0,
   $$
   entonces $x_0$ es un **mínimo local o relativo**.

4. **Máximo local:**  
   Si 
   $$
   f(x_0) \geq f(x), \quad \forall x \text{ cercano a } x_0,
   $$
   entonces $x_0$ es un **máximo local o relativo**.

### <u>Minimización y Maximización de Funciones Convexas</u>

#### Teorema

Sea $f$ una función **convexa** definida en un conjunto convexo no vacío $C$.

- El conjunto $A$ donde $f$ alcanza su **mínimo** es **convexo**.  
- Todo **mínimo local** de $f$ es también un **mínimo absoluto**.

### Punto Crítico

Sea $f : I \to \mathbb{R}$ una función continua y sea $x_0$ un punto en el **interior de $I$**.  

Decimos que $x_0$ es un **punto crítico** de $f$ si cumple al menos una de las siguientes condiciones:

1. $f'(x_0) = 0$  
2. $f$ **no es derivable** en $x_0$

### <u>Punto Crítico en Varias Variables</u>

#### Definición

Sea $f : A \to \mathbb{R}$ una función para la cual existen sus **derivadas parciales**.  

Decimos que $P$ es un **punto crítico** de $f$ si todas sus derivadas parciales se anulan en $P$, es decir:

$$
\frac{\partial f}{\partial x_i}(P) = 0, \quad \forall i = 1, \dots, n.
$$

Ejemplo:

Sea la función:

$$
f(x, y) = x^2 - 2x + 6y + y^2 + 10
$$

Para hallar los puntos críticos, resolvemos:

$$
\frac{\partial f}{\partial x} = 2x - 2 = 0, \quad \frac{\partial f}{\partial y} = 2y + 6 = 0
$$

De donde:

$$
x = 1, \quad y = -3
$$

Por lo tanto, $P = (1, -3)$ es un punto crítico de $f$.

#### Teorema

Sea $f : A \to \mathbb{R}$ una función para la cual existen sus derivadas parciales.  

- Si $f$ tiene un **máximo o mínimo local** en $P \in A$, entonces $P$ es un **punto crítico**.  
- El **recíproco** no siempre es cierto: un punto crítico **no garantiza** ser máximo o mínimo (puede ser un **punto de silla**).

### <u>Punto Silla</u>

Diremos que $P$ es un **punto de silla** de la función $f$ si es un **punto crítico**, pero **no es ni máximo ni mínimo local**.

Ejemplo:

Sea la función:

$$
f(x, y) = y^2 - x^2
$$

1. $(0, 0)$ es un **punto crítico**.  
2. $(0, 0)$ **no es máximo ni mínimo**, por lo que es un **punto de silla**.

### <u>Menores Principales de una Matriz Hessiana</u>

Dada la **matriz Hessiana** de $f$:

$$
H(f) =
\begin{bmatrix}
f_{x_1x_1} & f_{x_1x_2} & \cdots & f_{x_1x_n} \\
f_{x_2x_1} & f_{x_2x_2} & \cdots & f_{x_2x_n} \\
\vdots & \vdots & \ddots & \vdots \\
f_{x_nx_1} & f_{x_nx_2} & \cdots & f_{x_nx_n}
\end{bmatrix}
$$

Se definen las **matrices menores principales** como:

$$
H_1 = [ f_{x_1x_1} ],
\quad
H_2 =
\begin{bmatrix}
f_{x_1x_1} & f_{x_1x_2} \\
f_{x_2x_1} & f_{x_2x_2}
\end{bmatrix},
\quad
H_3 =
\begin{bmatrix}
f_{x_1x_1} & f_{x_1x_2} & f_{x_1x_3} \\
f_{x_2x_1} & f_{x_2x_2} & f_{x_2x_3} \\
f_{x_3x_1} & f_{x_3x_2} & f_{x_3x_3}
\end{bmatrix}, \;\; \dots
$$

Los determinantes correspondientes se llaman **menores principales**:

$$
D_1 = \det(H_1) = f_{xx}, \quad
D_2 = \det(H_2) = f_{xx}f_{yy} - f_{xy}f_{yx}, \quad
D_3 = \det(H_3), \;\; \dots
$$

### <u>Matriz Definida Positiva</u>

Las siguientes proposiciones son **equivalentes**:  

Una matriz $H(x)$ es **definida positiva** si se cumple alguna (y por equivalencia, todas) de las siguientes condiciones:

$$
x^T H(x) x > 0, \quad \forall x \in \mathbb{R}^n
$$

- Todos los autovalores de $H(x)$ son positivos.  
- Todos los determinantes de sus menores principales de $H(x)$ son positivos.

### <u>Matriz Definida Negativa</u>

Las siguientes proposiciones son **equivalentes**:  

Una matriz $H(x)$ es **definida negativa** si se cumple alguna (y por equivalencia, todas) de las siguientes condiciones:

$$
x^T H(x) x < 0, \quad \forall x \in \mathbb{R}^n
$$

- Todos los autovalores de $H(x)$ son negativos.  
- Los determinantes de los menores principales de $H(x)$ cumplen:  
  $$
  D_1 < 0, \quad D_2 > 0, \quad D_3 < 0, \quad D_4 > 0, \;\dots
  $$

### <u>Teorema: Extremos y la Matriz Hessiana</u>

Sea $f : \mathbb{R}^n \to \mathbb{R}$ una función cuya **segunda derivada es continua**, y sea $x_0$ un **punto crítico** de $f$.  

- Si en $x_0$ la **matriz Hessiana** es **definida negativa**, entonces $x_0$ es un **máximo**.  
- Si en $x_0$ la **matriz Hessiana** es **definida positiva**, entonces $x_0$ es un **mínimo**.

Ejemplo, sea la matriz Hessiana:

$$
H =
\begin{bmatrix}
x & xy \\
1 & x
\end{bmatrix}
$$

en el punto $x_0 = (-1, 1)$ se cumple que $$ D_1 = -1$$ y $$D_2=2$$ entonces $x_0$ es un máximo.

### <u>Hessiano para Funciones de Dos Variables</u>

Si $f$ es una función de dos variables, la **matriz Hessiana** de $f$ se reduce a:

$$
H(x, y) =
\begin{bmatrix}
f_{xx} & f_{xy} \\
f_{yx} & f_{yy}
\end{bmatrix}
$$

y su **determinante Hessiano** es:

$$
\Delta(x, y) = |H(x, y)| = f_{xx} f_{yy} - f_{xy}^2.
$$

### <u>Máximos y Mínimos de una Función de Dos Variables</u>

1. Hallamos los puntos críticos (posibles máximos o mínimos relativos).  
   Para esto resolvemos:

$$
f_x = 0, \quad f_y = 0
$$

y encontramos las soluciones \((x, y)\).

2. Para cada solución, aplicamos el siguiente teorema:

#### Teorema (Criterio de la Segunda Derivada)

Si $f(x, y)$ satisface las condiciones del **Teorema de Schwarz** y $(a, b)$ es un **punto crítico** de $f$:

a) Si 

$$
\Delta(a, b) > 0 \quad \text{y} \quad f_{xx}(a, b) > 0,
$$

entonces $(a, b)$ es un **mínimo local** de $f$.  

b) Si 

$$
\Delta(a, b) > 0 \quad \text{y} \quad f_{xx}(a, b) < 0,
$$

entonces $(a, b)$ es un **máximo local** de $f$.  

c) Si 

$$
\Delta(a, b) < 0,
$$

entonces $(a, b)$ es un **punto de silla** de $f$.  

d) En cualquier otro caso, el criterio **no es concluyente**.

### <u>Matriz Hessiana para Funciones de Tres Variables</u>

Dada la **matriz Hessiana** de $f$:

$$
H(f) =
\begin{bmatrix}
f_{xx} & f_{xy} & f_{xz} \\
f_{yx} & f_{yy} & f_{yz} \\
f_{zx} & f_{zy} & f_{zz}
\end{bmatrix},
$$

podemos extraer **3 sub-matrices** de la siguiente manera:

$$
H_1 = [ f_{xx} ],
\quad
H_2 =
\begin{bmatrix}
f_{xx} & f_{xy} \\
f_{yx} & f_{yy}
\end{bmatrix},
\quad
H_3 =
\begin{bmatrix}
f_{xx} & f_{xy} & f_{xz} \\
f_{yx} & f_{yy} & f_{yz} \\
f_{zx} & f_{zy} & f_{zz}
\end{bmatrix}.
$$

Los determinantes correspondientes se llaman **menores principales**:

$$
D_1 = \det(H_1) = f_{xx}, \quad
D_2 = \det(H_2) = f_{xx} f_{yy} - f_{xy} f_{yx}, \quad
D_3 = \det(H_3).
$$

### <u>Máximo y Mínimo de Funciones de Tres Variables</u>

1. Hallamos los puntos críticos (posibles máximos o mínimos relativos).  
   Para esto resolvemos:

$$
f_x = 0, \quad f_y = 0, \quad f_z = 0
$$

y encontramos las soluciones \((x, y, z)\).

2. Para cada solución, aplicamos el siguiente teorema:

#### Teorema (Criterio de la Segunda Derivada)

Sea $f : \mathbb{R}^3 \to \mathbb{R}$ una función cuyas **segundas derivadas parciales existen y son continuas**.  
Sea $(a, b, c)$ un **punto crítico** de $f$, y sean $D_1, D_2, D_3$ los **menores principales** de la matriz Hessiana evaluados en $(a, b, c)$:

a) Si 

$$
D_1(a, b, c) > 0, \quad D_2(a, b, c) > 0, \quad D_3(a, b, c) > 0,
$$

entonces $(a, b, c)$ es un **mínimo relativo** de $f$.  

b) Si 

$$
D_1(a, b, c) < 0, \quad D_2(a, b, c) > 0, \quad D_3(a, b, c) < 0,
$$

entonces $(a, b, c)$ es un **máximo relativo** de $f$.  

c) Si 

$$
D_1(a, b, c) \neq 0, \quad D_2(a, b, c) \neq 0, \quad D_3(a, b, c) \neq 0
$$

y no se cumplen los dos casos anteriores, entonces $(a, b, c)$ es un **punto de silla** de $f$.  

d) En cualquier otro caso, el criterio **no es concluyente**.
